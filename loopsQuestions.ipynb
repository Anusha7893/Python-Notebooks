```python
# Solution

first = int(input("Enter the start of range of numbers: "))
last = int(input("Enter the end for the range of numbers: "))
t = int(input("Enter the number to be divided with: "))

nums = []
nums1 = []

for i in range(first, last + 1):
  nums.append(i)

print("The numbers in range are: ", nums)
a = 0
for a in range(0, len(nums)):

    if nums[a] % t == 0:
        nums1.append(nums[a])
        a += 1

print(nums1)
total = 0
for a in range(0, len(nums1)):
    total += int(nums1[a])

print("Sum of all the numbers are: ", total)

```

```python
# Solution

num = int(input("Enter the number: "))
k = int(input("Enter the power to be raised: "))

series = []

for i in range(1, num+1):
  a = i ** k
  series.append(a)
  print(i)

print("\n")
print("The numbers are: ", series)
sum = 0
for a in range(0, len(series)):
    sum += int(series[a])

print("Sum of all the numbers are: ", sum)
```
