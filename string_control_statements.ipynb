#Extended Slicing and String Methods
##Extended Slicing
It is an extension of slicing that you have learnt.

**Syntax:**

`variable[start_index:end_index:step]`

 * Along with `start_index`, and `end_index` we will be providing `step`. The `step` determines the increment between each index for slicing.



```python
secret_message = "-R-a-v-i-"
print(secret_message[1:8:3])
secret_message1 = secret_message[1:8:3]
print(secret_message, secret_message1)
# Refer to image-1
```

```python
for i in range(1,11,2):
  print(i)
```

```python
a = "Waterfall"
part = a[1:6:3]
print(part)
# Refer to image-2
```

##Methods
Python has a set of built-in reusable utilities.

They simplify the most commonly performed operations are:

**String Methods**

* isdigit()
* strip()
* lower()
* upper()
* startswith()
* endswith()
* replace() and more...

## **Isdigit**
Syntax: `str_var.isdigit()`

Gives `True` if all the characters are digits. Otherwise, `False`  

```python
is_digit = "47a48afd".isdigit()
# print(is_digit)
num=""
for i in "47a48afd":
  if i.isdigit():
    num+=i
print(num)
```

## **Strip**
Syntax: `str_var.strip()`  

Removes all the leading and trailing spaces from a string.  

```python
mobile = " 9876543210 "
mobile = mobile.strip()
print(mobile)
```

## **Strip - Specific characters**
Syntax: `str_var.strip(chars)`  

We can also specify characters that need to be removed.

```python
name = ",Ravi.,"
name = name.strip(".,")
print(name)
```

## **Strip - Multiple Characters**
Removes all spaces, comma(,) and full stop(.) that lead or trail the string.  

```python
name = ", .. ,, ravi ,, .. ."
name = name.strip(" ,.")
print(name)
```

## **Replace**
Syntax: `str_var.replace(old,new)`  

Gives a new string after replacing all the occurrences of the old substring with the new substring.  

```python
sentence = "teh cat and teh dog"
modified_sentence = sentence.replace("teh","the")
print(modified_sentence)
```

## **Startswith**
Syntax: `str_var.startswith(value)`  

Gives `True` if the string starts with the specified value. Otherwise, `False`  

```python
url = "https://onthegomodel.com"
is_secure_url = url.startswith("https://")
print(is_secure_url)
```

## **Endswith**
Syntax: `str_var.endswith(value)`  

Gives `True` if the string ends with the specified value. Otherwise, `False`  

```python
gmail_id = "rahul123@gmail.com"
is_gmail = gmail_id.endswith("@gmail.com")
print(is_gmail)
```

## **Upper**
Syntax: `str_var.upper()`  

Gives a new string by converting each character of the given string to uppercase.  

```python
name = "Ravi"
print(name.upper())
```

## **Lower**
Syntax: `str_var.lower()`  

Gives a new string by converting each character of the given string to lowercase.  

```python
name = "RA$%VI"
print(name.lower())
```

**NOTE:**

The `upper()` and `lower()` methods in Python work only on **alphabetic characters** and on **special characters** (like punctuation marks, symbols, digits, etc.), these methods **do not have any effect**.

**For example:**

```python
s = "Hello, John! 123"
print(s.upper())  # Output: "HELLO, JOHN! 123"
print(s.lower())  # Output: "hello, john! 123"
```
